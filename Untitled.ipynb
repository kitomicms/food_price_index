```python
# find out about "Food index"

import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv("Food_price_indices_data_mar715.csv",skiprows=2,)
df = df.iloc[1:,0:7]
df['Date'] = df['Date'].apply(lambda x: datetime.strptime(x, '%Y-%m'))
df.set_index(['Date'],inplace=True)
```

```python
# latest date

df.index[-1]
```

```python
# columns name
df.columns
```

```python
# sub chart

fig, axs = plt.subplots(2, 3, sharex=True,figsize=(20,5))

axs[0, 0].plot(df['Meat'])
axs[0, 0].set_title('Meat')

axs[0, 1].plot(df['Dairy'], 'tab:orange')
axs[0, 1].set_title('Dairy')

axs[0, 2].plot(df['Cereals'], 'tab:green')
axs[0, 2].set_title('Cereals')

axs[1, 0].plot(df['Oils'], 'tab:red')
axs[1, 0].set_title('Oils')

axs[1, 1].plot(df['Sugar'], 'tab:purple')
axs[1, 1].set_title('Sugar')

axs[1, 2].plot(df['Food Price Index'], 'k') # https://matplotlib.org/3.5.1/tutorials/colors/colors.html
axs[1, 2].set_title('Food Price Index')
```
